```python
import os 
import pandas as pd 

export_dir = os.path.join('data', 'DASH_exports')
exports = [os.path.join(export_dir,fp) for fp in os.listdir(export_dir)]
```

```python
dataframes = []
careers = []
pot_duplicates = []
have_near_min = False 
for export in exports:
    df = pd.read_excel(export)
    if have_near_min: 
        df = df[df['CareerPath'] != 'Near-Minimum Wage Job'] 
    career = df['CareerPath'].iloc[0]
    if career in careers: 
        print(f"{os.path.basename(export)}\nAlready have {career}. Duplicate excel sheet?")
        pot_duplicates.append({'fp':os.path.basename(export),'career':career})
    else:
        careers.append(career) 
        dataframes.append(df)
        have_near_min = True
```

```python
df_concat = pd.concat(dataframes, axis=0)

# Are all the careers the same count? 
# Are the ages columns the same and consistent?
```
